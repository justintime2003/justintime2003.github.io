### NN Model Representation

Designed NN has three dense layers with sigmoid activations:
Images are 20 x 20 so we have 400 input features in $\vec{x}$

NN Architecture:
- 25 units in the first hidden layer, shape of W1 is (400, 25), shape of b is (25,)
- 15 units in the second hidden layer, shape of W2 is (25, 15), shape of b is (15,)
- 1 unit in the third layer (output layer), shape of W3 is (15, 1) and shape of b is (1,)


You specify a layer's output dimensions and this determines the next layer's input dimension. Input dimension of the first layer is derived from the size of the input data
```Python
tf.keras.Input(shape=(400,)), #specify input shape
```

### Exercise 1
Using Keras Sequential model and Dense Layer with sigmoid activation to constuct the NN described above

````python
model = Sequential(
    [
        tf.keras.Input(shape=(400,)),
        Dense(units=25, activation='sigmoid'),
        Dense(units=15, activation='sigmoid'),
        Dense(units=1, activation='sigmoid')
    ], name ="my_model"
)
````

````python
model.summary()
````

````python
#Number of parameters per layer
L1_num_params = 400 * 25 + 25  # W1 parameters  + b1 parameters
L2_num_params = 25 * 15 + 15   # W2 parameters  + b2 parameters
L3_num_params = 15 * 1 + 1     # W3 parameters  + b3 parameters
print("L1 params = ", L1_num_params, ", L2 params = ", L2_num_params, ",  L3 params = ", L3_num_params )
````

````python
[layer1, layer2, layer3] = model.layers
````

````python
#check the shape of the weights and bias for each layer
# xx.get_weights() returns a NumPy array
W1, b1 = layer1.get_weights()
W2, b2 = layer2.get_weights()
W3, b3 = layer3.get_weights()
print(f"W1 shape = {W1.shape}, b1 shape = {b1.shape}")
print(f"W2 shape = {W2.shape}, b2 shape = {b2.shape}")
print(f"W3 shape = {W3.shape}, b3 shape = {b3.shape}")
````

````python
print(model.layers[2].weights)
````

Define a loss function and run gradient descent to fit the weights of the model to training data,

````python
model.compile(
    loss=tf.keras.losses.BinaryCrossentropy(),
    optimizer=tf.keras.optimizers.legacy.Adam(0.001),
)
model.fit(X, y, epochs=20)
````

````python
prediction = model.predict(X[0].reshape(1,400))  # a zero
print(f" predicting a zero: {prediction}")
prediction = model.predict(X[500].reshape(1,400))  # a one
print(f" predicting a one:  {prediction}")
````

````python
if prediction >= 0.5:
    yhat = 1
else:
    yhat = 0
print("Predictions after threshold:", yhat)
````

````python
for i in range (500,1000,10):
    prediction = model.predict(X[i].reshape(1,400))
    print(f"prediction probability: {prediction}")
    if prediction >=0.5:
        yhat = 1
    else:
        yhat = 0
    print("Model Prediction: ", yhat)
````
